```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.Graph()
states = ['WA','NT','SA','Q','NSW','V']
colors=["red","blue","green"]
neighboures = dict()
```

```python
neighboures['WA'] =['NT','SA']
neighboures['NT'] =['WA','SA','Q']
neighboures['SA'] =['WA','NT','Q','NSW','V']
neighboures['Q']=['NT','NSW','SA']
neighboures['NSW']=['Q','SA','V']
neighboures['V']=['NSW','SA']
```

```python
color_of_state = dict()
for node in states:
  used_color= {color_of_state[neighbor] for neighbor in neighboures[node] if neighbor in color_of_state}
  for color in colors:
    if color not in used_color:
      color_of_state[node]=color
```

```python
color_of_node=[color_of_state[node] for node in states]
graphs=nx.Graph()
for node in states:
  for neighbor in neighboures[node]:
    graphs.add_edge(node,neighbor)
```

```python
nx.draw(graphs,node_color=color_of_node,with_labels=True)
```
